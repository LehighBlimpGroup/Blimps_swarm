```python
# Functions
'''
description : Generate a graph with random positions and save the data to Filename.txt
param        [*] Filename
param        [*] Graph_Size
param        [*] Number
return       [*] pos
'''
def Generate_Agents(Filename, Graph_Size, Number):
    # Generate nodes with random positions
    pos = {}
    for n in range(Number):
        pos[n] = (random.uniform(0, Graph_Size), random.uniform(0, Graph_Size))
    # Save the data to Filename.txt
    with open(Filename, 'w') as f:
        for key, value in pos.items():
            f.write('%s:%s\n' % (key, value))
    return pos

'''
description : Load the data from Filename.txt and generate a graph
param        [*] Filename
return       [*]
'''
def Load_Positives(Filename):
    # Create a graph with the data in Filename.txt
    G = nx.Graph()
    with open(Filename, 'r') as f:
        for line in f:
            node, pos = line.strip().split(':')
            G.add_node(node, pos=eval(pos))
    return G

def Draw_Graph(G):
    # Draw the graph
    pos = nx.get_node_attributes(G, 'pos')
    nx.draw(G, pos, with_labels=True, node_size=100, node_color='r')
    plt.show()
    
def 
```
